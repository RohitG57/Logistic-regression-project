```python
df2 = df[inte_data]
df2.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="mean")
impu = imputer.fit_transform(df2)
```

```python
df3 = pd.DataFrame(impu, columns=df2.columns,
                          index=df2.index)
```

```python
df3.isnull().sum()
```

```python
df4 = df[obj_data]
df4.head()
```

```python
def dummies(z,df):
    temp = pd.get_dummies(df[z], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([z], axis = 1, inplace = True)
    return df
# Applying the function to the cars_lr

df4 = dummies('age_bracket',df4)
df4 = dummies('qualification',df4)
df4 = dummies('race',df4)
df4 = dummies('sex',df4)
df4 = dummies('income_level',df4)
df4 = dummies('marital_status',df4)
df4 = dummies('employment',df4)
df4 = dummies('census_msa',df4)
```

```python
dataset = pd.concat([df3,df4],axis = 1)
data = pd.concat([dataset,df['h1n1_vaccine']],axis=1)
```

```python
data
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
x
```

```python
sns.distplot(y)
```

```python
y.hist(bins=50)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(solver = 'liblinear',max_iter=100)

# fit the model with data
logreg.fit(x_train,y_train)

#
y_pred=logreg.predict(x_test)
```

```python
logreg.score(x_train,y_train)
```

```python
model_score=logreg.score(x_test,y_test)
model_score
```

```python
# import the metrics class
from sklearn.metrics import confusion_matrix 
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import  matplotlib.pyplot as plt
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
y_pred_proba = logreg.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
